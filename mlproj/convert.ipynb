```python
# The 'heights' list is our input data, also known as 'features' or 'X'
heights = [[1.5], [1.6], [1.7], [1.8], [1.9]]

# The 'weights' list is our output data, also known as 'labels' or 'y'
weights = [[50], [60], [68], [75], [82]]

from sklearn.linear_model import LinearRegression

# Create a Linear Regression model object
model = LinearRegression()

# 'model.fit()' learns the relationship between heights and weights
model.fit(heights, weights)

# The new height we want to predict the weight for
new_height = [[2.4]]

# 'model.predict()' uses the learned relationship to make a prediction
predicted_weight = model.predict(new_height)

# The result is a NumPy array, so we can access the value with [0][0]
print(f"Predicted weight for a person of {new_height[0][0]:.2f}m: {predicted_weight[0][0]:.2f} kg")

import matplotlib.pyplot as plt

# Plot the original data points
plt.scatter(heights, weights, color='blue', label='Actual Data')

# Plot the regression line
plt.plot(heights, model.predict(heights), color='red', label='Regression Line')

# Add labels and a title
plt.title('Height vs. Weight Prediction')
plt.xlabel('Height (m)')
plt.ylabel('Weight (kg)')
plt.legend()
plt.grid(True)
plt.show()
```
